```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import tree

# you may get a warning from importing ensemble. It is OK to ignore said warning
from sklearn import ensemble
plt.style.use('fivethirtyeight')
```

```python
# Load US college data
us_college_data = pd.read_csv("social_capital_college.csv")
# us_college_data.head()
```

```python
# Load in county data
county_data = pd.read_csv("social_capital_county.csv")
# county_data.head()
```

```python
# Merge datasets
merged = us_college_data.merge(county_data, on="county")
```

```python
# Create new df that only contains colleges in California
california_colleges = merged.query('county_name.str.contains("California")', engine='python')
# california_colleges.head()
```

```python
def college_type(data):
    if "University of California-" in data:
        return "UC"
    elif "California State University-" in data:
        return "CSU"
    elif "Community College" in data:
        return "CCC"
    
california_colleges['college_type'] = california_colleges['college_name'].apply(college_type)
california_colleges['college_type'].value_counts()
```

```python
california_colleges = california_colleges.dropna()
```

```python
sns.scatterplot(data=california_colleges, x='ec_own_ses_college', y='mean_students_per_cohort', hue='college_type')
```

```python
ca_colleges_train, ca_colleges_test = train_test_split(california_colleges, test_size=0.25)
ca_colleges_train = ca_colleges_train.sort_values(by='college_type')
ca_colleges_test = ca_colleges_test.sort_values(by='college_type')
```

```python
logistic_regression_model = LogisticRegression(multi_class='ovr')
logistic_regression_model.fit(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
```

```python
log_regression_training_accuracy = logistic_regression_model.score(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
log_regression_test_accuracy = logistic_regression_model.score(ca_colleges_test[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_test['college_type'])
log_regression_training_accuracy, log_regression_test_accuracy
```

```python
decision_tree_model = tree.DecisionTreeClassifier(criterion='entropy', random_state=42)
decision_tree_model.fit(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
```

```python
dt_training_accuracy = decision_tree_model.score(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
dt_test_accuracy = decision_tree_model.score(ca_colleges_test[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_test['college_type'])
dt_training_accuracy, dt_test_accuracy
```

```python
random_forest_model = ensemble.RandomForestClassifier(criterion='entropy', random_state=42, n_estimators=20)
random_forest_model.fit(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
```

```python
rf_train_accuracy = random_forest_model.score(ca_colleges_train[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_train['college_type'])
rf_test_accuracy =  random_forest_model.score(ca_colleges_test[['ec_own_ses_college', 'mean_students_per_cohort']], ca_colleges_test['college_type'])
rf_train_accuracy, rf_test_accuracy
```

```python
def plot_centers_and_black_data(joined, centers):
    sns.scatterplot(data=joined, x='ec_own_ses_college', y='mean_students_per_cohort', color="black")
    for center in centers:
        plt.plot(center.coordinates[0], center.coordinates[1], '*', markersize = 10)    
    plt.xlabel('ec_own_ses_college')
    plt.ylabel('mean_students_per_cohort')
    legend_text = ['c' + str(i) for i in range(1, len(centers) + 1)]
    legend_text.append('data')
    plt.legend(legend_text)
```

```python
plot_centers_and_black_data(joined, (c1, c2, c3))
```
